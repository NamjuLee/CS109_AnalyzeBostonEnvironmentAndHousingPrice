```python
def SavePandasToCSV(path,d):
    d.to_csv(path)
    return "done!!"
def SaveJson(path, data): # "source\\K-town_streetviews.json"
    with open(path, 'w') as f:
        f.write(json.dumps(data, indent=4, ensure_ascii=False, encoding="latin1"))
        f.close()
def OpenJson(path): # "source\\K-town_streetviews.json"
    json_file = open(path , 'r')
    json_str = json_file.read()
    json_data = json.loads(json_str)
    return json_data
def GetGooglePlace(q, pos):
    YOUR_API_KEY =  'YOUR GOOGLE PLACE API!!!!' 
    google_places = GooglePlaces(YOUR_API_KEY)

    # You may prefer to use the text_search API, instead.
    '''
    query_result = google_places.nearby_search(
            location='berkeley, England', 
            keyword='apple',
            radius=2000, types=[types.TYPE_FOOD])
    '''
    Boston = {'lat':pos[1], 'lng':pos[0]} # {'lat':42.310052, 'lng':-71.079818}

    theKeyWord = q
    theType = "types.TYPE_"+theKeyWord.upper()

    query_result = google_places.radar_search(
            lat_lng = Boston,
            radius = 3200, #The radius (in meters) around the location/lat_lng to restrict the search to. The maximum is 50000 meters (default 3200)
            keyword = theKeyWord
            #types = [types.TYPE_DEPARTMENT_STORE] 
            )
    return query_result
```

```python
def GetPosFromQueryResult(QS):
    pos = []
    for j in QS.raw_response['results']:
        for i in j:
            if(i == "geometry"):
                pos.append({"lat":  float(j[i]['location']['lat']) , "long" : float(j[i]['location']['lng'])  } )
    print "the num of data: ",len(pos)
    return pos
```

```python
count =0;
output = {}

thePos = [ [-71.153433, 42.362948 ], [-71.07517, 42.362948 ], [-70.996442, 42.356785 ],
           [-71.07517, 42.30768 ], [-71.153433, 42.30768], [-71.153433, 42.252606], [-71.07517, 42.252606] ]
query = queryAll
for q in query:
    print q
    data = []
    try:
        for p in thePos:
            qs = GetGooglePlace(q, p)
            pos = GetPosFromQueryResult(qs)
            for temp in pos:
                data.append(temp)
            print count
            count += 1
        #     if(count > 3):break
        output[q] = data
    except:
        pass
print 'done'
print output
```
